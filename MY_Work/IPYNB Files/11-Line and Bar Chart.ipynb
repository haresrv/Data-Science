```python
x1=[1,3,5,7]
x2=[2,4,6,8]
y1=[7,11,8,7]
y2=[17,18,29,40]
plt.figure(figsize=(8,6))

ax=plt.axes()
ax.set_facecolor('white')
plt.bar(x1,y1,label="First",facecolor='#C0CA33',edgecolor='black')
plt.bar(x2,y2,label="Second")

for x,y in zip(x1,y1):
    plt.text(x,y+2,'%d'%y,ha='center',va='bottom')
for x,y in zip(x2,y2):
    plt.text(x,y+0.05,'%d'%y,ha='center',va='bottom')    

plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title('$Bar $ $ Chart$')
plt.legend()
plt.show()
```

```python
x1=[1,3,5,7]
x2=[2,4,6,8]
y1=[7,7,7,7]
y2=[17,18,29,40]
plt.figure(figsize=(8,6))

ax=plt.axes()
ax.set_facecolor('white')
plt.barh(x1,y1,label="First")
plt.barh(x1,y2,label="Second",left=y1)

for x,y in zip(x1,y1):
    plt.text(y-2,x,"%d" %y,ha='center',va='bottom')
for x,y,z in zip(x1,y2,y1):
    plt.text(y+z-10,x,"%d" %y,ha='center',va='bottom')    

plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title('$Bar $ $ Chart$')
plt.legend()
plt.show()
```
