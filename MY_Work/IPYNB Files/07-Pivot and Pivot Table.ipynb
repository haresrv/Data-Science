```python
import pandas as pd
import numpy as np
```

```python
d = pd.DataFrame({
    "Item": ['Item0', 'Item0', 'Item1', 'Item1'],
    "CType":['Gold', 'Bronze', 'Gold', 'Silver'],
    "USD":  ['1$', '2$', '3$', '4$'],
    "EU":  ['1€', '2€', '3€', '4€']
})
```

```python
d
```

```python
p = d.pivot(index='Item',columns='CType',values='USD')
p
```

```python
print(d[(d['CType']=='Gold') & (d['Item']=='Item0')].USD.values)
print(p[p.index=='Item0'].Gold.values)
```
