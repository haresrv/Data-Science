```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data', header = 0, na_values = ['?'], names=["symboling", "normalized_losses", "make", "fuel_type", "aspiration",
           "num_doors", "body_style", "drive_wheels", "engine_location",
           "wheel_base", "length", "width", "height", "curb_weight",
           "engine_type", "num_cylinders", "engine_size", "fuel_system",
           "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm",
           "city_mpg", "highway_mpg", "price"])
```

```python
df.dtypes
```

```python
obj_df = df.select_dtypes(include=['object']).copy()
obj_df.head()
```

```python
obj_df.isnull().sum()
```

```python
obj_df[obj_df.isnull().any(axis=1)]
```

```python
obj_df["num_doors"].value_counts()
```

```python
obj_df = obj_df.fillna({"num_doors":"four"})
```

```python
obj_df["num_cylinders"].value_counts()
```

```python
cleanup_nums = {"num_doors":{"four":4,"two":2},
            "num_cylinders":{"four":4,"two":2,"six":6,"five":5,"twelve":12,"three":3,"eight":8}}
```

```python
obj_df.replace(cleanup_nums,inplace=True)
obj_df.head()
```

```python
obj_df.dtypes
```

```python
obj_df["body_style"]=obj_df["body_style"].astype('category')
```

```python
obj_df["body_style_cat"]=obj_df["body_style"].cat.codes
obj_df.head()
```

```python
pd.get_dummies(obj_df,columns=["drive_wheels"]).head()
```

```python
pd.get_dummies(obj_df,columns=["body_style","drive_wheels"],prefix=["body","drive"]).head()
```

```python
obj_df["engine_type"].value_counts()
```

```python
obj_df["OHC_Code"] = np.where(obj_df["engine_type"].str.contains("ohc"),1,0)
```

```python
obj_df[["make","engine_type","OHC_Code"]].head()
```

```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
obj_df["make_code"] = lb_make.fit_transform(obj_df["make"])
obj_df[["make","make_code"]].head(19)
```

```python
from sklearn.preprocessing import LabelBinarizer

lb_style = LabelBinarizer()
lb_res = lb_style.fit_transform(obj_df["body_style"])
pd.DataFrame(lb_res,columns=lb_style.classes_).head()
```
