```python
plt.plot(x,x+0,linestyle='solid')
plt.plot(x,x+1,linestyle='dashed')
plt.plot(x,x+2,linestyle='dashdot')
plt.plot(x,x+3,linestyle='dotted')

plt.plot(x,x+4,linestyle='-')
plt.plot(x,x+5,linestyle='--') 
plt.plot(x,x+6,linestyle='-.')
plt.plot(x,x+7,linestyle=':')
```

```python
x = np.array([1,2,3,4,5,6])
y1 = x+1
y2 = x+3
y3 = x+4

plt.plot(x,y1,'b-',label='$y1=x+1$')
plt.plot(x,y2,'r^',label='$y2=x+2$')
plt.plot(x,y3,'gv',label='$y3=x+3$')

plt.xlabel("X - Axis")
plt.xlabel("Y - Axis")

plt.legend()
plt.tight_layout() #Automatically adjust sub plots
```

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x),label='$y1=x^2$')
plt.plot(x,np.cos(x),label='$y2=x^3$')

plt.xlabel(r'$X$',fontsize=18)
plt.xlabel(r'$Y$',fontsize=18)

plt.title('$Sin(x)$ $&$ $ Cos(x)$',fontsize=14)
plt.legend(loc='upper right')
plt.tight_layout() #Automatically adjust sub plots
```
