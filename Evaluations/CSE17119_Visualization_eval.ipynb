```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
columns = ['Age','Workclass','fnlgwt','Education','Education num','Marital Status',
           'Occupation','Relationship','Race','Sex','Capital Gain','Capital Loss',
           'Hours/Week','Native country','Income']
train = pd.read_csv('https://raw.githubusercontent.com/vivek2319/Predicting-US-Census-Income/master/Dataset/adult-training.csv', names=columns)
```

```python
train.head()
```

# Q1

```python
sns.scatterplot(y=train['Capital Gain'],x=train['Capital Loss'],size='Age' ,hue='Marital Status',data=train)
plt.legend(loc='upper right')
```

Conclusion

# Q2

```python
sns.lineplot(x="Age",y="Hours/Week",data=train ,marker='^',dashes=False, ci=68,color='green')
```

Conclusion

# Q3

```python
sns.stripplot(x = train.Age , y = train['Marital Status'])
sns.boxplot(x = train.Age , y = train['Marital Status'])
```

Conclusion

# Q4

```python
sns.violinplot(x="Relationship",y="Age",data=train,size=6)
sns.stripplot(x="Relationship",y="Age",data=train)
# sns.swarmplot(x="Relationship",y="Age",data=train) --> Takes long time
```

```python
sns.boxplot(y = train.Age , x = train['Marital Status'],hue=train['Sex'])
plt.legend(bbox_to_anchor=(1.3,0.5),loc='center right')
plt.xticks(rotation=45)
```

Conclusion
