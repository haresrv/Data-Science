```python
class display(object):
  template = """<div style="float: left; padding: 2px;">
  <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
  </div>"""
  def __init__(self, *args):
    self.args = args
    
  def _repr_html_(self):
    return '\n'.join(self.template.format(a, eval(a)._repr_html_())
             for a in self.args)
  
  def __repr__(self):
    return '\n\n'.join(a + '\n' + repr(eval(a))
              for a in self.args)
```

## 15. From the year columns delete the alphabets from the values and make it float

```python
df.columns
```

```python
df.loc[:,['2008 ','2009 ', '2010 ', '2011 ', '2012 ', '2013 ', '2014 ', '2015 ', '2016 ','2017 ', '2018 ', '2019']]
```

```python
import re
```

```python
year=['2008 ','2009 ', '2010 ', '2011 ', '2012 ', '2013 ', '2014 ', '2015 ', '2016 ','2017 ', '2018 ', '2019']
df.loc[:,year]=df.loc[:,year].applymap(lambda x: re.sub(r'[^0-9^\-\.]+', '', x) if not type(x)==float else x).replace('', np.float64(0)).astype('float64')
```

## 19. Sum of unemployment rate for male for EU27_2020 in year 2010

```python
melted = pd.melt(df,var_name='Year',value_name="Employment Rate",id_vars=['duration', 'isced11', 'age', 'sex', 'unit', 'geotime'])                  

melted.drop(['duration','isced11','age','unit'],axis=1,inplace=True)

melted['UnEmployment_Rate'] = melted['Employment Rate'].apply(lambda x:100-x)

melted.sample(20)
```

```python
ans = melted.groupby(['sex','Year']).sum()
ans.reset_index(inplace=True)
```

```python
ans.drop(['Employment Rate'],inplace=True,axis=1)
```

```python
ans[(ans['sex']=='M') & (ans['Year']=='2010 ')].set_index(['sex','Year'])
```
